```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import calendar
pd.options.display.max_columns = None
```

```python
df_covid = pd.read_csv('./Covid_clean.csv')
```

```python
df_covid.head()
```

```python

# Convertir la columna 'mes_nombre' a tipo 'category' con ordenamiento personalizado
df_covid['mes_nombre'] = pd.Categorical(df_covid['mes_nombre'], categories=calendar.month_name[1:], ordered=True)

# Agrupar y contar los contagios por mes_nombre
contagios_por_mes = df_covid['mes_nombre'].value_counts().sort_index()

# Crear el gráfico de barras
contagios_por_mes.plot(kind='bar', figsize=(10, 5))

# Título y etiquetas de los ejes
plt.title('Contagios por mes')
plt.xlabel('Mes')
plt.ylabel('Contagios')

# Mostrar el gráfico
plt.show()
```

```python
df_covid_fallecidos = df_covid[df_covid['fallecidos'] == 1]
```
